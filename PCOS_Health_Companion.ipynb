```python
import pandas as pd

# Reload datasets
pcos_df = pd.read_csv("PCOS_data.csv")
nutrition_df = pd.read_excel("nutrition.xlsx")
menstrual_df = pd.read_csv("periodCopy.csv")

# Remove the 'Unnamed: 44' column from PCOS dataset (it's empty)
if "Unnamed: 44" in pcos_df.columns:
    pcos_df.drop(columns=["Unnamed: 44"], inplace=True)

# Fill missing values in 'Marriage Status (Yrs)' and 'Fast food (Y/N)' with the most common value (mode)
pcos_df["Marraige Status (Yrs)"].fillna(pcos_df["Marraige Status (Yrs)"].mode()[0], inplace=True)
pcos_df["Fast food (Y/N)"].fillna(pcos_df["Fast food (Y/N)"].mode()[0], inplace=True)

# Check again for missing values
print("\nMissing values after cleaning in PCOS dataset:\n", pcos_df.isnull().sum())

# Save cleaned datasets for further use
pcos_df.to_csv("Cleaned_PCOS_data.csv", index=False)
```

```python
# Drop irrelevant columns
cols_to_drop = ["Sl. No", "Patient File No."]
pcos_df.drop(columns=cols_to_drop, inplace=True)

print("\nRemaining Columns in PCOS dataset:\n", pcos_df.columns)
```

```python
# Convert 'Y/N' columns to 1/0 (binary)
binary_cols = ["PCOS (Y/N)", "Pregnant(Y/N)", "Weight gain(Y/N)", "hair growth(Y/N)", 
               "Skin darkening (Y/N)", "Hair loss(Y/N)", "Pimples(Y/N)", "Fast food (Y/N)", 
               "Reg.Exercise(Y/N)"]

for col in binary_cols:
    pcos_df[col] = pcos_df[col].map({"Y": 1, "N": 0})

# Save the final cleaned dataset
pcos_df.to_csv("Final_PCOS_data.csv", index=False)

print("\nPCOS dataset cleaned and saved as 'Final_PCOS_data.csv'! ðŸš€")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

# Function to clean numeric values
def clean_numeric(value):
    try:
        return float(str(value).split()[0])  # Extract number part
    except ValueError:
        return np.nan  # Convert invalid values to NaN

# Apply cleaning function
nutrition_features = nutrition_features.applymap(clean_numeric)

# Debug: Check NaNs before dropping/filling
print("NaN count per column before processing:\n", nutrition_features.isna().sum())

# Fill NaN values instead of dropping
nutrition_features.fillna(nutrition_features.mean(), inplace=True)

# Debug: Check shape before scaling
print("Shape before scaling:", nutrition_features.shape)

# Apply StandardScaler
scaler = StandardScaler()
nutrition_scaled = scaler.fit_transform(nutrition_features)

# Convert back to DataFrame
nutrition_df = pd.DataFrame(nutrition_scaled, columns=nutrition_features.columns)

print("âœ… Nutrition dataset successfully scaled!")
```

```python
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters=3)  # Adjust clusters as needed
nutrition_hc_labels = hc.fit_predict(nutrition_scaled)
print(nutrition_hc_labels)
from sklearn.mixture import GaussianMixture

gmm = GaussianMixture(n_components=3)  # Adjust clusters as needed
gmm.fit(nutrition_scaled)
nutrition_gmm_labels = gmm.predict(nutrition_scaled)
nutrition_df['Cluster'] = nutrition_gmm_labels

# Ensure column names match the shape of nutrition_scaled
nutrition_df = pd.DataFrame(nutrition_scaled, columns=nutrition_features.columns)

# Assign cluster labels
nutrition_df['Cluster'] = nutrition_hc_labels  # Or nutrition_gmm_labels if using GMM

print(nutrition_df.head())  # Check the first few rows
```

```python
cluster_labels = {
    0: 'High-Protein Foods',
    1: 'Low-Carb & High-Fiber',
    2: 'Balanced Diet',
    3: 'High-Fat Foods'  # Adjust based on analysis
}

# Map numeric clusters to meaningful labels
nutrition_df['Cluster_Label'] = nutrition_df['Cluster'].map(cluster_labels)
print(nutrition_df[['Cluster', 'Cluster_Label']].head())  # Check the mapping
```

```python
def recommend_diet(cluster_label):
    if cluster_label == 'High-Protein Foods':
        return "Great for muscle building & weight management. Include in a balanced PCOS diet!"
    elif cluster_label == 'Low-Carb & High-Fiber':
        return "Excellent for blood sugar control & gut health. Ideal for PCOS management."
    elif cluster_label == 'Balanced Diet':
        return "Good for general well-being. Ensure portion control for PCOS."
    elif cluster_label == 'High-Fat Foods':
        return "Limit intake! Prioritize healthy fats for hormonal balance."
    else:
        return "No recommendation available."

# Apply function to dataset
nutrition_df['Diet_Recommendation'] = nutrition_df['Cluster_Label'].apply(recommend_diet)

# View sample recommendations
print(nutrition_df[['Cluster_Label', 'Diet_Recommendation']].drop_duplicates())
```
